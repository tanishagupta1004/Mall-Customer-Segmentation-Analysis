Clustering

```python
df1=df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
df1.head()
```

```python
#Scatter plot

sns.scatterplot(df1['Spending Score (1-100)'], df1['Annual Income (k$)'])
```

```python
from sklearn.cluster import KMeans
errors=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=1)
    kmeans.fit(df1)
    errors.append(kmeans.inertia_)
```

```python
#Plot the results for elbow method

plt.figure(figsize=(13,6))
plt.plot(range(1,11), errors)
plt.plot(range(1,11), errors, linewidth=3, color='red', marker='8')
plt.xlabel('No. of clusters')
plt.ylabel('WCSS')
plt.xticks(np.arange(1,11,1))
plt.show()
```

```python
km=KMeans(n_clusters=5)
km.fit(df1)
y=km.predict(df1)
df1['Label']=y
df1.head()
```

```python
sns.scatterplot(x='Annual Income (k$)',y='Spending Score (1-100)',data=df1,hue='Label',s=50,palette=['Red','Green','Blue','Purple','Orange'])
```
